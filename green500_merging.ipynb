```python
import pandas as pd

files = [
    'green500_top_202011.xlsx',
    'green500_top_202111.xlsx',
    'green500_top_202211.xlsx',
    'green500_top_202311.xlsx',
    'green500_top_202411.xlsx',
    'green500_top_202511.xlsx'
]

dfs = []

for file in files:
    year = file[13:17]
    df = pd.read_excel(file)
    df['source_year'] = int(year)
    dfs.append(df)
    print(f"Loaded {file}: {len(df)} rows")
combined_df = pd.concat(dfs, ignore_index=True)
combined_df.to_csv('green500_OG_combined.csv', index=False)
```
