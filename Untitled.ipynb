```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("merged_infrastructure_efficiency.csv")
df.info()
```

```python
sns.scatterplot(data = df, x = 'floor_space_sqft_total' , y = 'power_capacity_MW_total', hue = 'supercomputing_category')
plt.title('Power Capacity vs. Floor Space \n of Global Data Centers')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Floor Space (sqft)")
plt.ylabel("Power Capacity (MW)")
plt.grid(alpha=0.3)
plt.show()
```
